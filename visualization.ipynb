```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#features_pca = np.load('results/mnist_vgg11_pca.npy')
#features_tsne = np.load('results/mnist_vgg11_tsne.npy')

features_pca = np.load('results/fashion_vgg11_pca.npy')
features_tsne = np.load('results/fashion_vgg11_tsne.npy')
labels = np.load('results/fashion_labels.npy')[:10000]
```

```python
fashion_labels = {
    0: 'T-shirt/top',
    1: 'Trouser',
    2: 'Pullover',
    3: 'Dress',
    4: 'Coat',
    5: 'Sandal',
    6: 'Shirt',
    7: 'Sneaker',
    8: 'Bag',
    9: 'Ankle boot'
}
```

```python
palette = matplotlib.cm.get_cmap('tab10')
pca_pd = pd.DataFrame({'x1': features_pca[:, 0], 'x2': features_pca[:, 1], 'label': labels})
tsne_pd = pd.DataFrame({'x1': features_tsne[:, 0], 'x2': features_tsne[:, 1], 'label': labels})

pca_pd = pca_pd.replace({'label': fashion_labels})
tsne_pd = tsne_pd.replace({'label': fashion_labels})

fig, axes = plt.subplots(2, 1, figsize=(10, 20))
sns.scatterplot(x='x1', y='x2', hue='label', legend='full', data=pca_pd, ax=axes[0])
sns.scatterplot(x='x1', y='x2', hue='label', legend='full', data=tsne_pd, ax=axes[1])
```
